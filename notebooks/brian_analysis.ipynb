### Image loading


```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

img = Image.open("../data/synthetic_data/masks/synthetic_0000.tif")
frames = []

try:
    while True:
        frames.append(np.array(img))
        img.seek(img.tell() + 1)
except EOFError:
    pass  # No more frames

print(f"Number of frames extracted: {len(frames)}")
print(type(frames[0]))
print(frames[0].shape)


for frame in frames:
    plt.imshow(frame, cmap="gray")  # Remove cmap='gray' for color images
    plt.axis("off")
    plt.show()
```
