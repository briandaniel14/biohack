```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

img = Image.open("../data/ch20_URA7_URA8_001-crop1.tiff")
frames = []
try:
    while True:
        frames.append(np.array(img))
        img.seek(img.tell() + 1)
except EOFError:
    pass  # No more frames

print(f"Number of frames extracted: {len(frames)}")
print(type(frames[1]))

for frame in frames[3:10]:
    plt.imshow(frame, cmap="gray")  # Remove cmap='gray' for color images
    plt.axis("off")
    plt.show()
```
